Ejercicio_2

Dadas las listas A = ['a' , 'b' , '1' , '2' , 'e'] y B = ['1' , '2' , 'a' , 'b' , '5']. Es posible mutar una lista en un conjunto usando el comando set? En tal caso, calcule la unión, la intersección y la diferencia simétrica de los conjuntos resultantes. Es posible mutar un dato tipo conjunto a uno tipo lista? Investigue.

```python
A = ['a' , 'b' , '1' , '2' , 'e']
B = ['1' , '2' , 'a' , 'b' , '5']
a=set(A)
b=set(B)
c=a|b
d=a&b
e=a^b
print(c)
print(d)
print(e)
#Si es posible mutar un conjunto a una lista
C=list(c)
D=list(d)
E=list(e)
print(C)
print(D)
print(E)
```

Ejercicio_3

Diseñe un diccionario que contenga como valores, tres listas; la primera debe contener tres nombres de frutas, la segunda debe contener tres nombres de vegetales y la tercera tres nombres de granos. Elimine usando el diccionario construido el último elemento de cada lista de valores.


```python
dic={"frutas":["manzana","pera","banano"],"vegetales":["cebolla","pepino","lechuga"],"granos":["lenteja","arveja","frijol"]}
dic["frutas"].pop()
dic["vegetales"].pop()
dic["granos"].pop()
print(dic)
```

Ejercicio_5

Diseñe un diccionario que permita guardar 5 items (ej. autor, título, año, editorial, país) de un libro que escoja el usuario.

```python
libro={}
autor=str(input("Dime el autor de tu libro: "))
libro["autor"]=autor
titulo=str(input("Dime el titulo de tu libro: "))
libro["titulo"]=titulo
año=str(input("Dime el año de tu libro: "))
libro["año"]=año
editorial=str(input("Dime el editorial de tu libro: "))
libro["editorial"]=editorial
pais=str(input("Dime el pais de tu libro: "))
libro["pais"]=pais
print(libro)
```
