En este ejercicio se calcula el área de un círculo a partir del valor de su radio $r$. Nótese que asignamos un valor a $r$ que posteriormente se usa en la fórmula del cálculo del área del círculo.    `<variable> = (<valor_1>,<valor_2>,...,<valor_n>)`

   `<variable> = range(<valor_1>,<valor_2>,<paso>)`

Podemos agregar un elemento adicional a una lista mediante el comando` <lista>.append("<elemento>")`

Podemos eliminar un elemento de la lista, mediante el comando `<lista>.pop(<posición>)`

Pueden concatenarse dos listas para conformar una nueva usando el operador suma `+`

Podemos reversar una lista, usando el operador `<lista>.reverse()`

```python
x=["manzana","naranja","fresa"]
y=[2,3,5]
z=x+y
z.reverse()
print(z)
```

Podemos crear una lista de pares ordenados a partir de la composición de las funciones `list` y `zip` usando la sintaxis `list(zip(lista_1,lista_2, . . ., lista_n))`.

```python
x=["manzana","naranja","fresa"]
y=[2,3,5]
z=list(zip(x,y))
print(z)
```

### 1.3.5 Tipo diccionario
Un diccionario ` dict` , es una colección no ordenada de  items, que puede cambiar de tamaño. En un diccionario cada  item viene especificado por una ***clave*** o ***llave*** y un ***valor***. Llamaremos a un par `<clave i>:<valor i>` un ***item*** del diccionario.  La sintaxis para crear un diccionario es:

`<nombre-diccionario>={"<clave_1>" : "<valor_1>", ..., "<clave_n>" : "<valor_n>" }`,

también puede usarse:

 `dict(<clave_1>="<valor1>", ..., <clave_n>="<valor_n>" )`.

 **Nota**

1.   Un diccionario vacío se declara mediante el par de corchetes `{}`.
2.   El ***valor*** puede corresponder a una cadena de caracteres o a una lista.



**Ejemplo 1.**

Creamos una diccionario con nombre X, cuyas claves son: "fruta", "color" y "estado", y cuyos valores son: "manzana", "rojo" y "regular", respectivamente.

Creamos un diccionario con nombre `Digitos` conformado por digitos pares e impares.

```python
Digitos = {"pares":['0','2','4','6','8'], "impares":['1','3','5','7','9']}
print(Digitos)
```

1. El operador módulo `%`, retorna el resto que resulta al aplicar el algoritmo de la división a un par de enteros. Recordemos que, si tenemos dos números enteros  $m$ y $n$, con $n \neq 0$ y se efectua el algoritmo de la división, pueden enconstrarse dos enteros $q$ y $r$, tales que $$m = qn+r $$ en donde:
* si $n >0 $, entonces $0 \leq r < n$ y
* si $n < 0 $, entonces $n < r \leq 0$.

En esta división, $q$ se denomina *cociente* y $r$ *resto*. Por lo tanto, si efectuamos $m \% n$ obtendremos $r$. Observemos que $5 = 1*3 +2$, por lo tanto `5 % 3 = 2`, $6 = 2*3 + 0$ de donde `6 % 3 = 0`. Si $n=5$ y $m=-11$, observemos que $-11 = -3*5 +4$. Luego, `-11%5 = 4`. Si $n=-5$ y $m=-11$, observemos que $-11 = -2*5 - 1$. Luego, `-11%-5 = -1`.
2. El operador división piso `//`, retorna la parte entera que resulta de realizar una división entre dos enteros. Por lo tanto, si tenemos dos números enteros $m$ y $n$, con $n \neq 0$ y efectuamos el algoritmo de la división,  entonces el resultado de efectuar $m // n$ es $q$. Observemos que $5 = 1*3 +2$, por lo tanto `5 // 3 = 1`, de manera similar $6 = 2*3 + 0$ entonces`6 // 3 = 2`.

#2 Control de Flujo

Este tipo comandos, facilita el manejo del flujo lógico de las tareas o instrucciones que se realizan en un programa en Python.


## 2.1 Sentencia `if`
La declaración `if` permite trabajar en forma restringida un bloque de instrucciones, siempre y cuando una condición inicial se satisga o tome el valor `True`.

**Sintaxis**

`if` <**condición A**> `:`

         <instruccion 1>

         <instruccion 2>

         ...............

         <instruccion n>
**Nota**

Obsérvese que las instrucciones van indentadas respecto a la sentencia `if`,  esta indentación o pequeño espacio inicial es obligatorio, pues le indica al computador cuales instrucciones  deben realizarse, si la *condición A* toma el valor `True`. **Esta misma regla se mantiene para cualquier comando de control de flujo lógico**.

**Ejemplo 1.**

En el siguiente ejemplo, se le solicita al usario el nombre, si se llama "Juan" se  le saluda con un "Hola Juan", en otro caso se le ignora.

```python
n = input("Cuál es su nombre? ")
if n == "Juan":
   print("Hola Juan!")
```

La sentencia `for` permite ejecutar un número fijo de veces un bloque de instrucciones. Dicha sentencia se controla mediante una *variable* que toma valores de manera ordenada  sobre una secuencia; comenzando por el primero de sus valores. Cada vez que toma un valor la variable de control, se ejecuta el bloque de instrucciones. Una vez que ha tomado el último valor y ejecutado el bloque de instrucciones, el  ciclo `for` termina y se pasa a la siguiente instrucción del programa.

**Sintaxis**

for `<variable>` in `<secuencia>`:

       `<instrucción 1>`

       `<instrucción 2>`

        `..............`

       `<instrucción n>`

La serie de Fibonacci está definida por la fórmula recursiva

Diseñar una función que permita calcular la serie de Fibonaci, hasta un valor $n$ dado.

### 3.3.1 Uso de Lambda junto con los comandos  `filter`, `map` y `reduce`

La función lambda es muy útil para efectuar análisis de datos u operar  listas, en este caso haremos uso de las funciones `filter`, `map` y `reduce`. Estos comandos  permiten un diseño de códigos  versatiles y compactos.

* `filter` toma como argumento  un **conjunto** o una **secuencia**, y reporta una subcolección con los elementos del conjunto o la secuencia, que satisfacen una condición. La sintaxis es:
     
     `<nombre-lista> = list(filter(lambda <variable>: <condicion>,<secuencia>))`
* `map ` toma como argumentos una funciòn y un **conjunto** o una **secuencia**, y retorna una nueva lista con las evaluaciones de una función sobre cada elemento del conjunto o  la lista. La sintaxis puede presentarse de dos formas:

**(A)** Usando una funciòn definida previamente:

     `<nombre-lista> = list(map(<nombre_función>,<lista1>))`

**(B)** Definiendo una función mediante calculo lambda dentro del argumento de `map`:     
     `<nombre-lista> = list(map(lambda <variable>: <expresión>,<lista1>))`
* `reduce` toma como argumento una función  definida por `lambda`    y una **lista**, y retorna un valor que se calcula de manera (recursiva) acumulativa y ordenada sobre los elementos de la lista. En este caso debe usarse la librería `functools` e importar el comando `reduce`. La sintaxis es:

     `from functools import reduce`
     
     `<nombre-variable> = reduce(lambda <variable>: <expresión>,<lista1>)`

## 3.4 Variables Locales y Globales

Cuando se crea una función en Phyton,  se separa simultaneamente un espacio de la memoria para la función; esta área es de uso restringido de la función. En este caso, la variable definida dentro el cuerpo de instrucciones de la función se denomina **variable local**, y se dice que su **alcance** está restringido a la función.

En algunos casos es útil definir una variable como global para una función, en este caso dicha variable puede ser manipulada o cambiada por la función. Para que una variable en Phyton sea una variable global, es necesario que esté declarada como variable global dentro del cuerpo de instrucciones de alguna función. En este caso, sólo las funciones que tienen la variable definida como global, podrán manipular dicha variable. Si dicha variable global se encuentra fuera del cuerpo de la función, su valor podrá ser modificado.

**Ejemplo 1**

El siguiente ejemplo muestra el uso de las variables locales y globales, y el alcance de cada una de ellas. Nótese que la variable $c$ es local para las funciones $f2$ y $f3$, pero global para $f1$.

```python
# variables locales y globales
def f1():
    global c
    c = 13
    return(c)
def f2():
      global c
      c = 12
      return(c)
def f3():
      global c
      c = 10
      return(c)
# FIN DE LA FUNCION
# Intrucciones y variables externas a las funciones f1, f2 y f3.

c = 8
print ("\n Valor de la variable exterior c = ", c)
print ("\n Valor de la variable local c de la función f1 = ", f1())
print ("\n Valor de la variable exterior c = ", c)
print ("\n Valor de la variable local c de la función f2 = ", f2())
print ("\n Valor de la variable exterior c = ", c)
print ("\n Valor de la variable local c de la función f3 = ", f3())
print ("\n Valor de la variable exterior c = ", c)
## Ahora active la linea precedida con # en el código .... Verifique que sucede!!!!!
```

1. Diseñe una función que permita calcular la combinatoria
$C(n,k)=\frac{n!}{k!(n-k)!}$ Función combinatoria tiene un amplio uso en matemáticas, por ejemplo, genera los valores del Triángulo de Pascal, cuenta el número de subconjuntos con $k$ elementos que se pueden construir a partir de un conjunto con $n$ elementos, obviamente esta condición se aplica, si $k \leq n$.

```python
from functools import reduce
def factorial(x):
    l_x=[]
    for n in range(x):
        l_x.append(n+1)
    return reduce(lambda x,y: x*y, l_x)

def combninatoria (n,k):
    f_n=factorial(n)
    f_k=factorial(k)
    if k<n:
        return int(f_n/(f_k*factorial(n-k)))
    else:
        return "Eso no es posible"

print(combninatoria(9,3))
```

```python
n=1020302

def primo(num):
    if num == 1:
        return False
    for n in range(2, num//2 +1):
        if num % n == 0:
            return False
    return True

def Goldbach(nu):
    if nu % 2 != 0 or nu <= 2:
        return f"{nu} no cumple con las condiciones de la conjetura."
    cant=0
    for i in range (2,nu//2+1):
        if primo(i) and primo(nu - i):
            cant+=1
    return f"{nu} cumple con la conjetura al tener {cant} posibles sumas de primos que lo componen."

print(Goldbach(n))
```

# 4. Eficiencia de un algoritmo

La eficiencia de un algoritmo, puede medirse desde varios puntos de vista; por ejemplo, teniendo en cuenta la cantidad de recursos computacionales que se usan; como la cantidad de memoria requerida. También puede medirse calculando el tiempo requerido para resolver un problema. En esta sección, haremos una brevísima introducción al tema, con el fin de que el estudiante de un primer curso de programación, pueda entender la importancia del diseño eficiente de los algorimos. Esencialmente, mediremos la eficiencia contando el número de operaciones que realiza un algoritmo, esto es lo que comunmente se denomina complejidad de los algoritmos.





## 4.1 Ordenes de Complejidad

Una de manera de medir la complejidad de un algoritmo, consiste en calcular el número de operaciones que debe efectuar el algoritmo para resolver un problema. En algunos algoritmos, es posible encontrar   fórmulas que permiten estimar este número de operaciones.

Estudiaremos a continuación, el concepto de ordenes de complejidad, este concepto permite clasificar los algoritmos, de acuerdo a su grado de complejidad.

**Definición:**  Sea $f:\mathbb{N} \to \mathbb {R^{\geq 0}}$ una función. El conjunto de funciones de orden $f(n)$ es el conjunto definido por la igualdad

$$\mathcal O(f(n)) = \{g:\mathbb{N} \to \mathbb {R^{\geq 0}} | (\exists c \in \mathbb {R})(\exists N\in \mathbb{N}) (\forall n\geq N)( g(n) \leq cf(n))  \} $$

También diremos que una función $g(n)$ es de orden $f(n)$, si $g(n)\in \mathcal O(f(n))$

**Ejemplo 1.**

* $g_1(n) = n^3 +2n^2+1 \in \mathcal O(n^3)$
* $g_2(n) = \frac{n(n+1)}{2} \in \mathcal O(n^2)$
* $g_3(n) = 7n-22 \in \mathcal O(n)$
* $g_4(n) = 5\ln (n)+10 \in \mathcal O(\ln(n))$

**Definición:**  Diremos que $\mathcal O(g(n))$ es de orden menor o igual que $\mathcal O(f(n))$ y escribiremos $\mathcal O(g(n))\leq \mathcal O(f(n))$, si  $\mathcal O(g(n))\subseteq \mathcal O(f(n))$. Y escribiremos, $\mathcal O(g(n)) < \mathcal O(f(n))$, si  $\mathcal O(g(n))\subset \mathcal O(f(n))$.

**Ejemplo 2**

Nótese que,

$$ \mathcal O(x) \leq \mathcal O(5x+5) \leq \mathcal O(2x^2-5) \leq \mathcal O(x^2) $$
Además, observe que
$$\mathcal O(1) < \mathcal O(\sqrt n)< \mathcal O(\ln (n)) < \mathcal O(n) < \mathcal O(n\ln (n)) < \mathcal O(n^2) < \mathcal O(n^3) < \mathcal O(2^n) < \mathcal O(n!)$$

## 4.2 Complejidad de la Evaluación de Polinomios

Un polinomio es una expresión algebraica de la forma
$$p(x) = a_0 + a_1x + a_2x^2 + \cdots + a_{n-1}x^{n-1} + a_nx^n \text{       } (1)$$
donde cada $a_i$ es un número real o complejo y $x$ es una variable que puede tomar valores reales o complejos. Si hacemos un análisis del número de operaciones que se realizan al efectuar una evaluación del polinomio $p(x)$, usando la expresión dada arriba, veremos que:
* El número total de sumas es $n$.
* El número total de multiplicaciones es $\frac{n(n+1)}{2}$.

En algoritmos de tipo aritmético, se acostumbra a medir la complejidad, contando el número de multiplicaciones que se efectuan, pues esta operación exige más recursos y por lo tanto más tiempo de cálculo que una suma o una resta. Nótese, que si el polinomio es de grado $n=100$ efectuará $5050$ multiplicaciones, y si es de grado $1000$, efectuará $500500$ multiplicaciones.

La complejidad de un algoritmo numérico puede medirse por el orden del número de multiplicaciones que efectúe. Para el ejemplo anterior, observe que el orden de la complejidad del algoritmo sería $\mathcal O(n^2)$.


A continuación veremos un procedimiento que permite reducir el número de operaciones, y en consecuencia, será mucho más eficiente que el algoritmo que efectúa la evaluación realizando cada multiplicación y suma indicada en la expresión (1).

**El Algoritmo de Horner**

El polinomio dado anteriormente lo podemos escribir de la siguiente manera

$$p(x) = a_0 + a_1x + a_2x^2 + \cdots +a_{n-2}x^{n-2}+ (a_{n-1} + a_nx)x^{n-1}\\
= a_0 + a_1x + a_2x^2 + \cdots + a_{n-3}x^{n-3}  + (a_{n-2}+ (a_{n-1} + a_nx)x)x^{n-2}. $$

Continuando con este proceso algebraico, después de $n$ pasos obtendríamos,

$$p(x) = a_0 + (a_1 + (a_2 + \cdots   + (a_{n-2}+ (a_{n-1} + a_nx)x)...)x.$$

Observemos que hay que realizar la operación  $B+Cx$ varias veces, donde $B$ y $C$ toman los valores $a_{i-1}$ y $a_i$, respectivamente. Nótese que si se usa la última expresión para $p(x)$, podemos efectuar en cada paso una multiplicación y una suma, para obtener el resultado de la evaluación de $p(x)$. En total se efectuarán:

* $n$ sumas y
* $n$ multiplicaciones.

Así, si el polinomio es de grado $n=1000$, sólo se realizarían $1000$ multiplicaciones, siendo mucho más eficiente que el primero que describimos.
 Observe que el orden de la complejidad del algoritmo de Horner es $\mathcal O(n)$.
Veamos que aspecto tiene el algoritmo.

**Ejemplo 1**

Diseñe un algoritmo que permita evaluar cualquier polinomio por el método de Horner. Verifique su algoritmo con el polinomio $p(x) = 1 + 2x +3x^2$, calculando $p(1) = 6$.

```python
# Algoritmo de Horner para la evaluación de un polinómio p(x).
# Datos de entrada:
#   1.  A = lista de coeficiente del polinomio p(x).
#   2.  x = valor en donde se calcula el polinomio p(x).
# Datos de salida:
#       p(x) = Evaluación del polinomio en x.

#       DEFINICIÓN DE LA FUNCIÓN DE EVALUACIÓN    #

def EvalHorn(A,x):
    l = len(A)   # longitud de la lista A
    n = l-1      # grado del polinomio
    B = A[n-1]
    C = A[n]
    Eval = B + C*x
    for i in range(n-2,-1,-1):
        B = A[i]
        C = Eval
        Eval = B + C*x
    print(f"\n p({x}) = {Eval}")
#              LLAMADO A LA FUNCIÓN               #
EvalHorn([1,1,1,1,1],2)
```

A continuación mostramos una segunda versión del mismo algoritmo que no usa las variables auxiliares $B$ y $C$.

```python
def EvalHorn(A,x):
    n = len(A)-1      # grado del polinomio
    Eval = A[n]
    for i in range(n-1,-1,-1):
        Eval = A[i-1] + Eval*x
    print(f"\n p({x}) = {Eval}")
#              LLAMADO A LA FUNCIÓN               #
EvalHorn([1,1,1,1,1],2)
```

## 4.3 Complejidad de la combinatoria $C(n,k)=\frac{n!}{k!(n-k)!}$

Para calcular el valor $C(n,k)$, con $k \leq n$, necesitamos calcular tres valores previamente: $n!$, $k!$ y $(n-k)!$. Ahora, cada uno de estos factoriales necesita efectuar $n-1$, $k-1$, y  $n-k-1$ productos, respectivamente para determinar cada uno de los resultados. Así, en principio, el algoritmo necesitaría un total de $n-1+k-1+n-k-1 = 2n-3$ multiplicaciones, es decir, tendría una  complejidad  de orden $\mathcal O(2n)$. Sin embargo, la complejidad la podríamos reducir observando que $k!$ y $(n-k)!$, pueden obtenerse del cálculo de $n!$. Un algoritmo que implemente esta última observación tendría una complejidad de orden $\mathcal O(n)$.

**Ejemplo 1**

Escribir un algoritmo con complejidad $\mathcal O(n)$ que permita calcular los números combinatorios.


```python
def C(n,k):
  F = 1
  c = 1
  A = 1
  B = 1
  i = 2
  while i in range(2,n+1):
      if k == n or k == 0:
        c = 1
        i = n + 1
      else:
        F *=i
        if i == k:
          A = F
        if i == n-k:
          B = F
        i +=1
      c=F/(A*B)
  return c
C(5,3)
```

## 4.4 Complejidad del ordenamiento de una lista de números

Supongamos que tenemos una lista de $n$ números que queremos ordenar en forma creciente. Para calcular la complejidad computacional de este algoritmo, contamos el número de comparaciones necesarias para ordenar la lista. Nótese que el algoritmo está constituido por dos `for ` anidados. En el primer `for`, la variable $i$ toma $n-1$ valores posibles, por cada uno de estos valores, la varible $j$ del segundo `for ` toma $n-i$ valores posibles. Por lo tanto el número de comparaciones es $n-1 + n-2 + \cdots + 1 = \frac{n(n-1)}{2}=\frac{1}{2}(n^2-n)$. Es decir el algoritmo es de orden $\cal O(\frac{1}{2}n^2)$.

```python
def Ordenar(L):
  n = len(L)
  for i in range(0,n-1):
    max = L[i]
    for j in range(i+1,n):
      if max < L[j]:
         max = L[j]
         L[j] = L[i]
         L[i] = max
L = [20,6,3,12,5,2,10,11]
Ordenar(L)
print(L)
```

Calcule la complejidad de su algoritmo en términos del número de comparaciones necesarias para ordenar las listas de $n$.


```python
def Ordenar(L):
    for h in range(0,len(L)):
        n = len(L[h])
        for i in range(0,n-1):
            min = L[h][i]
            for j in range(i+1,n):
                if min > L[h][j]:
                    min = L[h][j]
                    L[h][j] = L[h][i]
                    L[h][i] = min

L = [[20,6,3,12,5,2,10,11],[215,2185,1354,1,54,8,6,14],[54,84,51,84,56,988,325,458,7]]
Ordenar(L)
print(L)
```

Para calcular la complejidad computacional de este algoritmo, asumamos tener una lista L con $n$ listas, cada una con $n$ datos, contamos el número de comparaciones necesarias para ordenar las lista. Nótese que el algoritmo está constituido por tre `for` anidados. En el primer `for`, la variable $h$ toma tantos valores posibles como listas en la lista L ($m$), por cada uno de estos valores, en el segundo `for`, la variable $i$ toma $n-1$ valores posibles, por cada uno de estos valores, la varible $j$ del tercer `for` toma $n-i$ valores posibles. Por lo tanto el número de comparaciones es $m \times (n-1 + n-2 + \cdots + 1) = m \times (\frac{n(n-1)}{2})=m \times \frac{1}{2}(n^2-n)$. Es decir el algoritmo es de orden $\cal O(\frac{1}{2}m \times n^2)$.

Escribir una algoritmo con complejidad de orden $\mathcal O(\frac{1}{2}n)$ que calcule los números combinatorios $C(n,k)$. **Ayuda:** Los números combinatorios pueden obtenerse construyendo el triángulo de Pascal ( https://es.wikipedia.org/wiki/Tri%C3%A1ngulo_de_Pascal ). Observe la simetría que tiene esta estructura y usela en el diseño de su algoritmo.


```python
def C(n, k):
    if k > n-k:
        k=n-k
    c=1
    for i in range(k):
        c*=(n - i)
        c//=(i + 1)
    return c

n=5
k=3
print(f"La combinatoria de {n} y {k} es {C(n,k)}")
```

6. Construya un algoritmo que calcule el mayor número de una lista conformada por $n$ números. Determine la complejidad del algoritmo.

```python
def maxi(L):
    n = len(L)
    max=0
    for i in range(0,n-1):
        if max < L[i]:
            max=L[i]
    return max

L=[2, 3, 5, 6, 10, 11, 12, 20, 1, 6, 8, 14, 54, 215, 1354, 2185, 7, 51, 54, 56, 84, 84, 325, 458, 988]
print(f"El número maximo de la lista {L} es {maxi(L)}")
```

## 5.1 Creación o Apertura de Archivos

La creación y apertura de un archivo, se realiza mediante una variable apuntadora. Esta variable guardará la dirección de la posición del primer dato del archivo. Dado que los archivos son datos secuenciales, es decir la información guardada se encuentra en posiciones sucesivas de memoria, con la dirección del primer elemento del archivo será fácil determinar la dirección de los elementos subsiguientes y accederlos. Este mecanismo de acceso es invisible para el programador, sin embargo, es importante entenderlo.

La creación o apertura de un archivo se realiza mediante el comando `open()`. La sintaxis de uso de este comando es:

* `<apuntador_archivo> = open(<"ruta_de_acceso_al_archivo">,<modo_acceso_archivo>)`

donde:

* `<apuntador_archivo>` es una variable de tipo apuntador que guarda la dirección de la primera posición del archivo.

* `<ruta_de_acceso_al_archivo>`  corresponde a la dirección  o ruta en donde se encuentra ubicado el archivo.
* `<modo_acceso_archivo>` corresponde a uno de los siguientes modos:

1. 'r' abre un archivo de solo **lectura**, iniciando en la posición `0` del archivo. Si el archivo referenciado no existe se reportará un error de tipo `FileNotFoundError`.
2. 'w' abre un archivo de solo **escritura**.  Si el archivo ya existe, el contenido de éste será borrado y reemplazado.
3. 'x' abre un archivo de solo **escritura**. Si el archivo referenciado existe se reportará un error de tipo `FileExistsError`.
4. 'r+' abre un archivo de **lectura** y **escritura**. Si el archivo referenciado no existe se reportará un error de tipo FileNotFoundError.
5. 'w+' abre un archivo de **lectura** y **escritura**. Si el archivo ya existe, el contenido de este será borrado y reemplazado.

## 5.3 Comandos para manejo de archivos

* `close()` permite cerrar en forma apropiada archivos. Este comando es de uso recomendado cuando se abre un archivo en modo escritura, simpre es necesario cerrar  los archivos, pues podrían sobreescribirse y dañarse en caso de que quede abierto. La sintaxis es
`<apuntador_archivo>.close()`.
* `read()` Permite leer un archivo que esté en modo de lectura `r`. Puede ingresarse en su argumento la posición hasta   la cual se quiere leer, recordando que la posición inicial comienza en `0` y que los caracteres de espaciado y fin de línea también hacen parte de cada línea.
 La sintaxis es `<apuntador_archivo>.read(<numero_posicion>)`.

* `write()` Permite escribir sobre un archivo que esté en modo de escritura` w`. Se escribirá  la cadena de caracteres que se indique en su argumento. Cada vez que se usa el comando `write()` se efectuan dos procesos: primero se escribe un caracter sobre la dirección apuntada y posteriormente el apuntador se modifica y guarda la siguiente posición  del archivo. La sintaxis es:  `<apuntador_archivo>.write(<cadena de caracteres>)`.

* `readline()` Permite leer una línea de un archivo comenzando por la primera, lo cual quiere decir que lee toda una cadena de caracteres hasta que encuentre el `caracter de fin de línea` denotado por `\n`. * `readlines()` Permite leer todas las lineas del archivo y conformar una lista de lineas.

**Formatos de impresión para write()**

Debe tenerse en cuenta que el comando `write()` funciona en forma ligeramente distinta que el comando `print()`; en el primero, debe indicarse el formato de impresión según el tipo de dato a escribir:

* Para `int` debe usarse `%nd`, donde $n$ corresponde al número de espacios que se separan para la impresión. Por ejemplo, si se usa el formato `%5d` para imprimir 25, quedaría `---25`.

* Para `float` debe usarse `%p.qf`, donde $p$ es el número de espacios que se dejan para escribir el número, de estos espacios se separan q para la parte decimal incluyendo el punto decimal. Por ejemplo, si se usa %8.3f, para escribir el número  57.327, se imprimirá ---57.32
